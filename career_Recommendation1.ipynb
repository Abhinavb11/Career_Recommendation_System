```python
from scipy.stats import chi2_contingency

# Chi-square test for each feature vs. Career
for col in df.columns[:-1]:  # Exclude Career column
    contingency_table = pd.crosstab(df[col], df["Courses"])
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    print(f"Feature: {col}, p-value: {p}")

    # If p-value < 0.05, the feature is useful
    if p < 0.05:
        print(f"âœ… {col} is relevant")
    else:
        print(f"âŒ {col} may not be useful")
```
